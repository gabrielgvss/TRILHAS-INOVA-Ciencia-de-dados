```python
# IMPORTANDO BIBLIOTECA NUMPY
import numpy as np
import matplotlib.pyplot as plt
```

```python
# IMPORTANDO DADOS DA WEB COM LOADTXT (fonte, delimitador do arquivo, colunas a serem utilizadas)
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'
dados = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 88, 1))

dados
```

```python
# VERIFICANDO NÚMERO DE DIMENSÕES DOS DADOS
dados.ndim #2 -> matriz (linhas/colunas)
```

```python
# VERIFICANDO QUANTIDADE DE ELEMENTOS DENTRO DA MATRIZ
print(f"Quantidade de linhas: {dados.shape[0]}\nQuantidade de colunas: {dados.shape[1]}")
print(f"Quantidade total de elementos na matriz: {dados.size}")
```

```python
# REALIZANDO TRANSPOSIÇÃO (TROCA DE LINHAS POR COLUNAS) PARA VISUALIZAÇÃO DOS DADOS
matriz_transposta = dados.T
matriz_transposta
```

```python
# SEPARANDO DADOS
datas = matriz_transposta[:, 0] #todas as linhas da primeira coluna
precos = matriz_transposta[:, 1:6] #todas as linhas da segunda até a sexta coluna
```

```python
# TRATANDO VALORES DE DATAS
datas = np.arange(1,88,1)
```

```python
# PLOTANDO RESULTADOS DA CIDADE DE MOSKOW
plt.plot(datas, precos[:, 0])
```

```python
# DEFININDO CIDADES DOS DADOS
Moscow = precos[:, 0]
Kaliningrad = precos[:, 1]
Petersburg = precos[:, 2]
Krasnodar = precos[:, 3]
Ekaterinburg = precos[:, 4]
```
