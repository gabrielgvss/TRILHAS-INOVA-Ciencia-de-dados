* Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a média de emissão de cada atividade econômica no ano de 2021.

```python
from numpy import array
# Registrando colunas com dados descritivos e dados temporais
colunas_desc = dados_gee.loc[:, 'Nível 1 - Setor':'Produto'].columns

colunas_temporais = dados_gee.loc[:, 1970:2021].columns
colunas_temporais = array(colunas_temporais)

# Aplicando melt para mudar o formato tabular de wide para long
dados_gee = dados_gee.melt(id_vars=colunas_desc, value_vars=colunas_temporais, var_name='Ano', value_name='Emissão')
dados_gee
```

```python
# Filtrando por emissão
dados_emissao2021 = dados_gee.query('`Emissão / Remoção / Bunker` == "Emissão" and `Ano` == 2021')
```

```python
# Agrupando por média de emissão por atividade econômica
media_emissao_setor = dados_emissao2021.groupby(['Nível 1 - Setor', 'Ano'])[['Emissão']].mean()

# Tratamento de colunas
media_emissao_setor['Emissão'] = round(media_emissao_setor['Emissão'], 2)
media_emissao_setor.rename(columns={'Emissão':'Média de Emissão de Gases'}, inplace=True)

media_emissao_setor.sort_values(by='Média de Emissão de Gases', ascending=False)
```

* Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a soma de emissão de cada atividade econômica. Ordene os dados da maior para menor emissão.

```python
# Selecionando dados de emissão de gases
dados_emissao = dados_gee.query('`Emissão / Remoção / Bunker` == "Emissão"')
del dados_emissao['Emissão / Remoção / Bunker']

# Agrupando total de emissão de gases por setor
soma_emissao_setor = dados_emissao.groupby('Nível 1 - Setor')[['Emissão']].sum()

# Tratando colunas
soma_emissao_setor.rename(columns={'Emissão': 'Total em Emissões'}, inplace=True)

# Arredondando a coluna 'Total em Emissões' para 2 casas decimais
soma_emissao_setor['Total em Emissões'] = soma_emissao_setor['Total em Emissões'].round(2)

# Ordenando os valores
soma_emissao_setor = soma_emissao_setor.sort_values(by='Total em Emissões', ascending=False)

# Formatando para notação decimal padrão
pd.options.display.float_format = '{:,.2f}'.format

soma_emissao_setor
```

- A equipe de supervisão deseja saber agora a média de emissão para cada um dos setores econômicos da coluna "Nível 1 - Setor". Como desafio, crie uma tabela de pivô e uma visualização gráfica contendo essas informações, utilizando apenas o método pivot_table para a construção da tabela, sem utilizar o método groupby.

```python
dados_emissao.columns

emissao_setor = dados_emissao.pivot_table(values='Emissão', index='Ano', columns='Nível 1 - Setor', aggfunc='mean')

emissao_setor
```

```python
emissao_setor.plot(subplots = True, figsize = (10,10));
```
